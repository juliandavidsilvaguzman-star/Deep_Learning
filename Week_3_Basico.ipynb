# Neurona Artificial Básica



## Objetivos de aprendizaje
- Calcular la **suma ponderada** y la **salida** de una neurona con función **escalón** (step).
- Medir el **error** respecto a una salida esperada.
- **Actualizar** el **peso** y el **bias** con una regla de aprendizaje muy simple.
- Observar cómo el error disminuye y cómo se comporta la neurona al final del entrenamiento.


```python
# ============================
# 1) Importar librerías
# ============================
# Solo usamos NumPy para arreglos y operaciones básicas.
import numpy as np
```

## Datos de ejemplo
Usaremos un conjunto mínimo de ejemplos para que la neurona **imite** el comportamiento identitario (entrada 0 → salida 0, entrada 1 → salida 1).

```python
# Entradas (x) y salidas esperadas (y)
# Dos ejemplos muy simples: 0->0 y 1->1
x = np.array([0, 1])
y = np.array([0, 1])

print('Entradas:', x)
print('Salidas esperadas:', y)
```

## Función de activación (Step / Escalón)
La función **step** devuelve 1 si la suma ponderada es **mayor o igual** a 0, y 0 en caso contrario. Esto nos permite obtener salidas **discretas** (0 o 1).

```python
# ============================
# 2) Definir la función de activación
# ============================

def step(z: float) -> int:
    """Función escalón (step):
    - Entrada: z (suma ponderada)
    - Salida: 1 si z >= 0; 0 si z < 0
    """
    return 1 if z >= 0 else 0

# Prueba rápida de la función
for val in [-1.0, 0.0, 0.5]:
    print(f"step({val}) = {step(val)}")
```

## Inicialización de parámetros
- **w**: peso inicial de la neurona
- **b**: bias (sesgo) inicial
- **lr**: tasa de aprendizaje (qué tan grandes son los ajustes en cada paso)

> Nota: Usamos valores sencillos para **ver** claramente los cambios.


```python
# ============================
# 3) Inicializar peso, bias y tasa de aprendizaje
# ============================

w = 0.5   # peso inicial
b = 0.0   # bias inicial
lr = 0.1  # tasa de aprendizaje (learning rate)

print(f"Peso inicial (w): {w}")
print(f"Bias inicial (b): {b}")
print(f"Tasa de aprendizaje (lr): {lr}")
```

**texto en negrita**## Entrenamiento (regla de actualización muy simple)
Para cada ejemplo:
1. Calculamos la **suma ponderada** `z = x[i] * w + b`.
2. Obtenemos la **salida** `y_pred = step(z)`.
3. Calculamos el **error** `error = y[i] - y_pred`.
4. **Actualizamos** los parámetros:
   - `w = w + lr * error * x[i]`
   - `b = b + lr * error`

> Intuición: si la neurona **se equivoca**, movemos `w` y `b` un poquito (proporcional a `lr`) en la dirección que **reduce el error**.


```python
# ============================
# 4) Bucle de entrenamiento
# ============================

# Número de pasadas por los datos (épocas)
epochs = 5

for epoch in range(1, epochs + 1):
    print(f"\nÉpoca {epoch}")

    for i in range(len(x)):
        # 1) Suma ponderada
        z = x[i] * w + b

        # 2) Salida de la neurona (0 o 1)
        y_pred = step(z)

        # 3) Error (qué tanto nos equivocamos)
        error = y[i] - y_pred

        # 4) Actualización (regla muy simple)
        w = w + lr * error * x[i]
        b = b + lr * error

        print(f"Entrada: {x[i]} | z: {z:.3f} | y_pred: {y_pred} | y_real: {y[i]} | error: {error} | w: {w:.3f} | b: {b:.3f}")

print("\nParámetros finales:")
print(f"w = {w:.3f}, b = {b:.3f}")
```

## Salida
Verificamos la salida de la neurona con los mismos ejemplos de entrenamiento para confirmar qué aprendió.

```python
# ============================
# 5) Evaluación
# ============================
print("\nResultados finales:")
for i in range(len(x)):
    z = x[i] * w + b
    y_pred = step(z)
    print(f"Entrada: {x[i]} → Salida predicha: {y_pred} (z={z:.3f}) | Esperada: {y[i]}")
```

## 📝 Conclusiones breves
- La neurona **ajusta w y b** para reducir los errores en sus predicciones.
- La función **step** permite obtener salidas binarias (0/1) fáciles de interpretar.
- Este mecanismo básico es la **base** de algoritmos más avanzados (p. ej., perceptrón y backpropagation en redes profundas).
